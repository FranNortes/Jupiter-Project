## Cambio de divisas

Gracias a nuestros asesores, nos dimos cuenta que el análisis de salarios no podría funcionar, ya que teníamos diferentes monedas y algunos valores eran superiores solo por el hecho de la moneda en la que estaba.



Creamos una función que busca los tipo de cambios de monedas a EUR, utilizando una API gratuita que genera los tipo de cambio y creamos otra función que ejecuta este cambio

```python
import requests

def get_exchange_rates():
    url = "https://api.exchangerate-api.com/v4/latest/EUR"
    response = requests.get(url)
    data = response.json()
    return data.get("rates", {})

def actualizar_tasas_conversion():
    tasas_conversion = {
        'USD': 0.0,
        'SGD': 0.0,
        'GBP': 0.0,
        'INR': 0.0,
        'MXN': 0.0,
        'IRR': 0.0,
        'EUR': 1.00
    }

    tasas_actualizadas = get_exchange_rates()

    for moneda in tasas_conversion.keys():
        if moneda in tasas_actualizadas:
            tasas_conversion[moneda] = tasas_actualizadas[moneda]

    return tasas_conversion

tasas_conversion = actualizar_tasas_conversion()
print(tasas_conversion)
```

```python
def convertir_a_euros(df, Salario_Minimo, Salario_Maximo, columna_moneda):
    for index, row in df.iterrows():
        moneda = row[columna_moneda]
        if moneda in tasas_conversion:
            tasa = tasas_conversion[moneda]
            # Convertir salarios a euros
            df.at[index, Salario_Minimo] = row[Salario_Minimo] / tasa
            df.at[index, Salario_Maximo] = row[Salario_Maximo] / tasa
            # Cambiar la moneda a 'EUR'
            df.at[index, columna_moneda] = 'EUR'

    return df

headhunter_final = convertir_a_euros(headhunter, 'Salario_Minimo', 'Salario_Maximo', 'Moneda')

headhunter_final[['Salario_Minimo', 'Salario_Maximo', 'Moneda']].head
```
